```python
def lower(st):
    newi = st
    count = 251
    ret_df = pd.DataFrame()
    while True:
        
        pre = newi[:newi.find('a)')].strip()
        quest = pre[pre.rfind('\n') + 1:]
        # print(pre)

        ans1 = newi[newi.find('a)'):newi.find('b)')]
        ans2 = newi[newi.find('b)'):newi.find('c)')]
        ans3 = newi[newi.find('c)'):newi.find('d)')]
        prob = newi[newi.find('d)'):].find('\n')
        ans4 = newi[newi.find('d)'):newi.find('d)') + prob]

        aft = newi[newi.find('d)') + prob:].find(')') + 1
        smth = len(newi[:newi.find('d)') + prob]) + aft - 1

        ans = newi[smth - 1]
        newi = newi[smth + 1:]
        tmp = pd.DataFrame()
        tmp['topic'] = [topics[count - 1]]
        tmp['question'] = [quest.strip()]
        tmp['option_1'] = ans1.strip()
        tmp['option_2'] = ans2.strip()
        tmp['option_3'] = ans3.strip()
        tmp['option_4'] = ans4.strip()
        if ans == 'a':
            tmp['answer'] = '1'
        if ans == 'b':
            tmp['answer'] = '2'
        if ans == 'c':
            tmp['answer'] = '3'
        if ans == 'd':
            tmp['answer'] = '4'
        ret_df = pd.concat([ret_df, tmp], axis=0)
        count += 1
        if count % 10 == 1:
            break
    return ret_df
```
